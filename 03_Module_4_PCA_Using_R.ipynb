```r
install.packages("readxl") # Install if not already installed
library(readxl)            # Load the package
```

```r
# Drag & Drop the Health file into files

myData <- read_excel("/content/Health.xlsx")
```

```r
# scale function to standardize the variables (excluding the first variable,
# Country Name) and store the standardized values

myData.st <- scale(myData[ ,-1])
```

```r
# use the prcomp function to perform PCA and use the
# summary function to display the PCA results

pca <- prcomp(myData.st)
summary(pca)
```

```r
# To review the weights for computing the principal components,
# we display the rotation property of the PCA results

pca$rotation
```

```r
# To review the principal component scores, we display the x
# property of the PCA results

pca$x
```

```r
# In order to list countries (first variable) along
# with the principal component scores

newData <- data.frame(myData, pca$x)
newData <- newData[ , -(2:13)]
head(newData)
```
